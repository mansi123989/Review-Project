```python
import re
def remove_pun(word):
    return re.sub(r'[^\W\S]' , ' ' , word)
```

```python
df = df.rename(columns = {'text_':'reviewDescription'})
```

```python
df['reviewDescription'][40427]
```

```python
df.reviewDescription = df.reviewDescription.apply(remove_pun)
```

```python
import nltk

from nltk.tokenize import word_tokenize
```

```python
from nltk.corpus import stopwords
stop_word = set(stopwords.words('english'))

def remove_stop_word(review):
    words = word_tokenize(review.lower())
    filtered = [word for word in words if word.isalpha() and word not in stop_word]
    return ' '.join(filtered)

df.review_stop_word = df.reviewDescription.apply(remove_stop_word)
```

```python
from nltk.stem import PorterStemmer 
stemmer = PorterStemmer()

def stem_word(text):
    words = text.split()
    stemmed = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed)

df['root_review'] =  df.review_stop_word.apply(stem_word)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()


```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x = df[['root_review']].squeeze()
y = df[['label']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(
    steps=[
        ('tfidf',CountVectorizer() ),
        ('ir', MultinomialNB())
    ]
)
```

```python

pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)

from sklearn.metrics import accuracy_score

print("",accuracy_score(y_test,y_pred))
```
